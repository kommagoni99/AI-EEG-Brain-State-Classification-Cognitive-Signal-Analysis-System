```python
# ==========================================================
# AI EEG BRAIN STATE CLASSIFICATION SYSTEM
# ==========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.fft import fft
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

np.random.seed(42)

# ----------------------------------------------------------
# 1. SIMULATE EEG SIGNALS
# ----------------------------------------------------------

def generate_eeg(state, length=256):
    t = np.linspace(0, 1, length)

    if state == "Relaxed":      # Alpha dominant (8-12 Hz)
        signal = np.sin(2 * np.pi * 10 * t)
    elif state == "Focused":    # Beta dominant (13-30 Hz)
        signal = np.sin(2 * np.pi * 20 * t)
    elif state == "Stressed":   # Gamma dominant (30-45 Hz)
        signal = np.sin(2 * np.pi * 35 * t)

    noise = np.random.normal(0, 0.5, length)
    return signal + noise

# ----------------------------------------------------------
# 2. DATASET CREATION
# ----------------------------------------------------------

states = ["Relaxed", "Focused", "Stressed"]
data = []
labels = []

for _ in range(300):
    for state in states:
        eeg = generate_eeg(state)
        fft_values = np.abs(fft(eeg))[:50]
        data.append(fft_values)
        labels.append(state)

X = np.array(data)
y = np.array(labels)

# ----------------------------------------------------------
# 3. TRAIN TEST SPLIT
# ----------------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# ----------------------------------------------------------
# 4. MODEL TRAINING
# ----------------------------------------------------------

model = RandomForestClassifier(n_estimators=150)
model.fit(X_train, y_train)

# ----------------------------------------------------------
# 5. EVALUATION
# ----------------------------------------------------------

preds = model.predict(X_test)

print("\nClassification Report:\n")
print(classification_report(y_test, preds))

cm = confusion_matrix(y_test, preds)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, cmap="Blues", fmt='d')
plt.title("Brain State Classification Confusion Matrix")
plt.show()

# ----------------------------------------------------------
# 6. SIGNAL VISUALIZATION
# ----------------------------------------------------------

example_signal = generate_eeg("Focused")

plt.figure(figsize=(10,4))
plt.plot(example_signal)
plt.title("Sample EEG Signal (Focused State)")
plt.show()

# ----------------------------------------------------------
# 7. LIVE BRAIN STATE PREDICTOR
# ----------------------------------------------------------

print("\n------ LIVE BRAIN STATE PREDICTOR ------")
print("Simulating new EEG signal...")

test_signal = generate_eeg(np.random.choice(states))
fft_test = np.abs(fft(test_signal))[:50].reshape(1, -1)

prediction = model.predict(fft_test)[0]

plt.figure(figsize=(10,4))
plt.plot(test_signal)
plt.title("New EEG Signal")
plt.show()

print("Predicted Brain State:", prediction)
```
